## Analysis

```python
import pandas as pd
```

```python
raw_df = pd.read_csv(r"C:\Users\r04ra18\Desktop\fault-dataset\Execl_files_work\Fault-number-analysis.csv")
```

```python
horizons_ls = raw_df['Horizon'].values.tolist()
```

```python
data_df = raw_df.iloc[:-1,1:].astype(float).copy().fillna(0)
```

```python
data_df.insert(0, 'Horizon', horizons_ls[:-1])
```

```python
horizons_df = pd.read_csv(r'C:\Users\r04ra18\Desktop\fault-dataset\Execl_files_work\after-reviwers\Hrs_names.csv')
```

```python
result_df = pd.merge(horizons_df, data_df, how="left", on=["Horizon"])
```

```python
result_df.describe()
```

## Section-1

```python
faults_section_1_df = result_df[['Horizon', 'Proven-1', 'Presumed-1']]
```

```python
ax = faults_section_1_df.plot.barh(x='Horizon', stacked=True, figsize=(10, 8), 
                                   xlabel='Horizons', ylabel='Section-12 Total Horizons Length (m)',
                                   color =['brown', 'lightpink', 'cyan'])

ax.set_title('Total Number of Faults - Section-1', weight='bold', fontsize=16)
ax.invert_yaxis()
# Set x-axis label
ax.set_xlabel("Number of Faults", weight='normal', size=12) # , labelpad=20

# Set y-axis label
ax.set_ylabel("Coal Seams Names", weight='normal', size=12);
```

```python
boxplot = faults_section_1_df.boxplot(column=['Proven-1', 'Presumed-1']) 
```

```python
faults_section_1_df_throw = result_df[['Horizon', 'throw_>_50m_1', 'throw_<_50m_1']]
```

```python
ax = faults_section_1_df_throw.plot.barh(x='Horizon', stacked=True, figsize=(10, 8), 
                                   xlabel='Horizons',
                                   color =['Black', 'cyan'])

ax.set_title('Total Number of Faults - Section-1', weight='bold', fontsize=16)
ax.invert_yaxis()
# Set x-axis label
ax.set_xlabel("Number of Faults", weight='normal', size=12) # , labelpad=20

# Set y-axis label
ax.set_ylabel("Coal Seams Names", weight='normal', size=12);
```
